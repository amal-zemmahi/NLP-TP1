```python
##   TP1 using Spacy librery

##   English text
###  import librery 

import spacy
nlp = spacy.load("en_core_web_sm")

#loading data
text="""Perhaps one of the most significant advances made by Arabic mathematics began at this time with the work of al-Khwarizmi, namely
the beginnings of algebra. It is important to understand just how significant this new idea was. It was a revolutionary move away from
the Greek concept of mathematics which was essentially geometry. Algebra was a unifying theory which allowed rational
numbers, irrational numbers, geometrical magnitudes, etc., to all be treated as "algebraic objects". It gave mathematics a whole new
development path so much broader in concept to that which had existed before, and provided a vehicle for future development of the
subject. Another important aspect of the introduction of algebraic ideas was that it allowed mathematics to be applied to itself in a
way which had not happened before."""

file=nlp(text)
```

```python
# Tokenization and Pos tagging

for token in file:
    print(token,'-->',(token.pos_),'-->',(token.tag_))
```

```python
# Lemmatization:

for word in file :
    print(word.text + '-->', word.lemma_)
```

```python
# Stemming using NLTK librery

from nltk.stem.porter import *
p_stemmer = PorterStemmer()
for words in file:
    print (words.text + ' --> ' + p_stemmer.stem(words.text))
```

```python
# Parsing

for token in file:
    print(token,token.dep_)
#print the index of current token ,the token,the token itself,the dependency,the dependency, the head and its index
    print(token.i,token,token.dep_,token.head.i,token.head)
```

```python
# Arabic version

import spacy 
from spacy.lang.en.stop_words import STOP_WORDS
from spacy.lang.ar import Arabic

# Loading Data
nlp= Arabic()
text_ar="ربما كانت أحد أهم التطورات التي قامت بها الرياضيات العربية التي بدأت في هذا الوقت بعمل الخوارزمي و هي بدايات الجبر، و من المهم فهم كيف كانت هذه الفكرة الجديدة مهمة، فقد كانت خطوة ثورية بعيدا عن المفهوم اليوناني للرياضيات التي هي في جوهرها هندسة، الجبر كان نظرية موحدة تتيح الأعداد الكسرية و الأعداد اللا كسرية، و قدم وسيلة للتنمية في هذا الموضوع مستقبلا. و جانب آخر مهم لإدخال أفكار الجبر و هو أنه سمح بتطبيق الرياضيات على نفسها بطريقة لم تحدث من قبل"
file = nlp(text_ar)

# Tokenization
token_words= [token.text for token in file]
print(token_words)
```

```python
# Lemmatization:

lemma= [[token.text,token.lemma_] for token in file]
print(lemma)
```

```python
# Pos Tagging

for word in file:
    print(word," : ",word.pos_," : ",word.tag_)
```

```python
# Steamming 
# using NLTK librery

from nltk.stem.porter import *
p_stemmer = PorterStemmer()
for words in file:
    print (words.text + ' --> ' + p_stemmer.stem(words.text))
```
